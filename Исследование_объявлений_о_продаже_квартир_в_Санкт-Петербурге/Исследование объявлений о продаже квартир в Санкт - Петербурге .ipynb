### Категоризация этажей

Напишем функцию floor которая будет перебирать значения в столбцах `floors_total` и `floor` и будет присваивать категории (первый, последний, другой) соответствующим записям:

```python
def floor(data):
    if data['floor'] == 1:
        return 'первый'
    elif data['floor'] == data['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

Выполним функцию и результат ее работы сохраним в новом столбце `floor_category`.

```python
df['floor_category']=df.apply(floor, axis = 1)
```

```python
df_1 = df.query('floors_total == floor')
df_2 = df.query('floor == 1')
df_3 = df.query('floors_total != floor')
```
